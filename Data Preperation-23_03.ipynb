# Reduce Large Categories

```python
df1.info()
```

```python
# Reduce large category sets in categorical variables
categorical_cols = ['playlist_genre', 'playlist_subgenre', 'track_artist']

# Count category frequencies for each column
category_counts = {col: df1[col].value_counts() for col in categorical_cols}

# Define general threshold: less than 1% of total rows
threshold = len(df1) * 0.01

# Create mapping: group rare categories into "Other"
category_mappings = {}

for col in categorical_cols:
    if col == 'track_artist':
        # Only group artists that appear exactly once
        rare = category_counts[col][category_counts[col] == 1].index.tolist()
    else:
        # For other columns, group categories with frequency < 1% of total rows
        rare = category_counts[col][category_counts[col] < threshold].index.tolist()

    category_mappings[col] = {cat: "Other" for cat in rare}

# Apply replacements to the DataFrame
df2=df1.copy()
for col, mapping in category_mappings.items():
    df2[col] = df2[col].replace(mapping)
    
# Show the number of unique values before/after for each column
before_after = {
    col: {
        "before": df1[col].nunique(),
        "after": df2[col].nunique()
    }
    for col in categorical_cols
}

# Display the updated DataFrame

display(before_after)
```

### `playlist_subgenre`

```python
# Define the manual mapping from subgenre to broader genre families
subgenre_family_map = {
    # Hip-Hop / Rap
    'southern hip hop': 'Hip-Hop / Rap',
    'latin hip hop': 'Hip-Hop / Rap',
    'gangster rap': 'Hip-Hop / Rap',
    'hip hop': 'Hip-Hop / Rap',
    'trap': 'Hip-Hop / Rap',

    # Pop
    'indie poptimism': 'Pop',
    'dance pop': 'Pop',
    'latin pop': 'Pop',
    'hip pop': 'Pop',
    'post-teen pop': 'Pop',

    # Rock / Classic Rock
    'hard rock': 'Rock',
    'urban contemporary': 'Rock',
    'classic rock': 'Rock',
    'album rock': 'Rock',

    # Electronic / EDM
    'progressive electro house': 'Electronic',
    'pop edm': 'Electronic',
    'electro house': 'Electronic',
    'big room': 'Electronic',
    'electropop': 'Electronic',

    # Alternative / Indie
    'neo soul': 'Alternative / Indie',
    'new jack swing': 'Alternative / Indie',
    'permanent wave': 'Alternative / Indie',

    # Tropical / World
    'tropical': 'Tropical / World',
    'reggaeton': 'Tropical / World'
}

# Create a new column 'genre_family' based on the mapping
df2['subgenre_family'] = df2['playlist_subgenre'].map(subgenre_family_map).fillna('Other')

# Display unique values and a few rows to confirm
df_family_summary = df2[['playlist_subgenre', 'subgenre_family']].drop_duplicates().sort_values(by='subgenre_family')

display(df_family_summary)
```

```python
# One-hot encode the new 'subgenre_family' column
subgenre_family_dummies = pd.get_dummies(df2['subgenre_family'], prefix='subgenre')

# Merge the dummies into the main dataframe
df2 = pd.concat([df2, subgenre_family_dummies], axis=1)

# Drop the original 'playlist_subgenre' and 'genre_family' if needed
df2 = df2.drop(columns=['playlist_subgenre', 'subgenre_family'])

# Show a few rows to confirm the encoding
display(df2)



```

# Album release date

An obvious yet important data preparation step is to convert dates stored as object (string) to type datetime. In our dataset the track_album_release_date column is a relevant case. Lets change this column to be type datetime:

```python
# Convert to datetime, handling cases with only a year
def convert_dates(date):
    if len(date) == 4:  # If it's just a year
        return pd.to_datetime(date + '-01-01')  # Assume January 1st as the default
    return pd.to_datetime(date, errors='coerce')  # Convert normally, setting invalid values to NaT

df2['track_album_release_date'] = df2['track_album_release_date'].apply(convert_dates)
```
